```python
import csv
from bs4 import BeautifulSoup
from IPython.display import SVG
```

```python
svg =open('./svg/Korea_black.svg','r').read()
```

```python
dust = csv.reader(open('./data/dust.csv','r'),delimiter=",")
```

```python
datadict = {}
```

```python
for row in dust :
    district = row[0]
    count =float(row[3])
    datadict[district] = count
datadict
```

```python
soup = BeautifulSoup(svg,"lxml")
paths = soup.find_all('g')
paths
```

```python
colors = ["#ffffb2","#fed976","#feb24c","#fd8d3c","#f03b20","#bd0026"]
```

```python
for p in paths:
    if p['id']:
        count = datadict[p['id']]
    if count >=39 :
        color_class =5
    elif count >=37:
        color_class =4
    elif count >=35:
        color_class =3
    elif count >=33:
        color_class =2
    elif count >=31:
        color_class =1
    else :
        color_class =0
    color = colors[color_class]
    p['fill'] = color
```

```python
svg_result = soup.prettify()
SVG(svg_result)
```
